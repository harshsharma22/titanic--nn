```
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import Binarizer
from sklearn.preprocessing import MinMaxScaler

def normalize_age(data):

    scaler = MinMaxScaler()
    data["Age"] = scaler.fit_transform(data["Age"].values.reshape(-1,1))
    return data

def binar(data):

    binarizer = Binarizer(0.5)
    test_predict_result = binarizer.fit_transform(data)
    test_predict_result = test_predict_result.astype(np.int32)
    return test_predict_result

def process(data):

    gender = LabelEncoder()
    gender.fit(['male', 'female'])
# goes through each column in data['sex'] and transform it into a 0 or a 1 and replace it in the origional position
    data["Sex"] = gender.transform(data["Sex"])
    ab =LabelEncoder()
    ab.fit(["S", "Q", "C"])
    data['Embarked'] = ab.transform(data['Embarked'])
    return data


def col(data):

    drop = ['PassengerId', 'Name', 'Ticket', 'Cabin','Fare']
    for col in drop:
        data = data.drop(col, axis=1)

    return data


traning_data = pd.read_csv("train.csv")
traning_data = traning_data.dropna()
y_traning = traning_data[['Survived']].values
print(y_traning.shape)

x_traning = traning_data.drop('Survived', axis=1)
x_traning = col(x_traning)
#print(x_traning.head(10))
x_traning = process(x_traning)
#print(x_traning.head(10))
x_traning = normalize_age(x_traning)

test_data= pd.read_csv("test.csv")

test_data = process(test_data)

test_data = col(test_data)
test_data=test_data.fillna(0)
test_data = normalize_age(test_data)

learning_rate = 0.003
traning_epochs = 10000
display_step = 5

number_of_input = 6
number_of_output = 1

layer_1_node = 18
layer_2_node = 36
layer_3_node = 18

with tf.variable_scope('input'):
    x = tf.placeholder(tf.float32, shape=(None,number_of_input))

with tf.variable_scope('layer1'):
    weight1 = tf.get_variable(name="weight1", shape=[number_of_input,layer_1_node],initializer= tf.contrib.layers.xavier_initializer())
    biases1= tf.get_variable(name="biases1",shape=[layer_1_node],initializer= tf.zeros_initializer)
    layer_1_output = tf.nn.sigmoid(tf.matmul(x, weight1)+biases1)

with tf.variable_scope('layer2'):
    weight2 = tf.get_variable(name="weight2", shape=[layer_1_node,layer_2_node],initializer= tf.contrib.layers.xavier_initializer())
    biases2 = tf.get_variable(name="biases2",shape=[layer_2_node], initializer= tf.zeros_initializer)
    layer_2_output = tf.nn.tanh(tf.matmul(layer_1_output, weight2)+biases2)

with tf.variable_scope('layer3'):
    weight3 = tf.get_variable(name="weight3", shape=[layer_2_node,layer_3_node],initializer= tf.contrib.layers.xavier_initializer())
    biases3 = tf.get_variable(name="biases3",shape=[layer_3_node],initializer= tf.zeros_initializer)
    layer_3_output = tf.nn.sigmoid(tf.matmul(layer_2_output, weight3)+biases3)

with tf.variable_scope('output'):
    weight4 = tf.get_variable(name="weight3", shape=[layer_3_node,number_of_output],initializer=tf.contrib.layers.xavier_initializer())
    biases4 = tf.get_variable(name="biases3", shape=[number_of_output], initializer=tf.zeros_initializer)
    predection = tf.nn.tanh(tf.matmul(layer_3_output, weight4) + biases4)

with tf.variable_scope('cost'):
    Y = tf.placeholder(tf.float32, shape=[183, 1])
    cost = tf.reduce_mean(tf.squared_difference(predection,y_traning))

with tf.variable_scope('train'):
    optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost)


with tf.Session() as session:

    session.run(tf.global_variables_initializer())

    for epoch in range(traning_epochs):
        session.run(optimizer, feed_dict={x: x_traning, Y: y_traning})
        if epoch %5==0:
            t=session.run(cost ,feed_dict={x:x_traning,Y: y_traning})
            print(epoch,t)
    print("traning complete")

    f_t = session.run(cost, feed_dict={x: x_traning, Y: y_traning})
    gt = session.run(predection, feed_dict={x: test_data})

    gt = binar(gt)
    print(gt[:10],y_traning[:10])
    test_data_1 = pd.read_csv("test.csv")
    passengerid = test_data_1["PassengerId"].to_frame()
    passengerid["Survived"] = gt
    passengerid.to_csv("predection1.csv",index=False)
    print("done")
    #test_passenger["Survived"] = gt
```
